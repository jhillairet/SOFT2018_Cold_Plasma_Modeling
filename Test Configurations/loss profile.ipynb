```python
import numpy as np
import matplotlib.pylab as plt
```

```python
X_max = 2*700e-3
Z_max = 2*700e-3

x_plasma = np.linspace(0, X_max, 101, endpoint=True)
z_plasma = np.linspace(0, Z_max, 101, endpoint=True)

losses = np.logspace(-3, 1, 6, endpoint=True)
```

```python
fig,ax=plt.subplots()
for l in losses:
    eps_sec = -1 + (1 - np.exp(x_plasma*l))
    ax.plot(x_plasma, abs(eps_sec))
    
ax.set_yscale('log')
ax.grid(True)
ax.grid(True, 'minor'      )
```

```python
_losses = 5
X, Z = np.meshgrid(x_plasma, z_plasma)

C = 0.1
L = -C*(1 + 
        (X>700e-3)*(-1+np.exp((X-700e-3)*_losses)) + 
        (Z>=3*Z_max/4)*(-1+np.exp((Z-3*Z_max/4)*_losses)) + 
        (Z<=Z_max/4)*(-1+np.exp(-(Z-Z_max/4)*_losses))) 
```

```python
fig,ax = plt.subplots()
ax.plot(x_plasma, -C*(1 + (x_plasma>700e-3)*(-1+np.exp((x_plasma-700e-3)*_losses))))
```

```python
fig,ax=plt.subplots()
con = ax.pcolor(X, Z, L)
cb = fig.colorbar(con)
ax.axis('equal')

```
