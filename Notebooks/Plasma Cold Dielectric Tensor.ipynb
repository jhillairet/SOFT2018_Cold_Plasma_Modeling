# Cold Plasma Dielectric parameters

The cold plasma dielectric tensor is defined as :
$$
\mathbf{K}
=
\left(
\begin{array}{ccc}
S & j D & 0 \\
-j D & S & 0 \\
0 & 0 & P
\end{array}
\right)
$$
where
$$
\begin{eqnarray}
S &=& 1 - \sum_s \frac{\omega_{p,s}^2}{\omega^2 - \Omega_{c,s}^2} \\
D &=& \sum_s \frac{\Omega_{c,s}}{\omega} \frac{\omega_{p,s}^2}{\omega^2 - \Omega_{c,s}^2} \\
P &=& 1 - \sum_s \frac{\omega_{p,s}^2}{\omega^2}
\end{eqnarray}
$$

where $\omega$ is the RF angular frequency ($\omega=2\pi f$) and $\Omega_{c,s}$, $\omega_{p,s}$ are the cyclotron and the plasma  angular frequencies respectively for the plasma species $s$.

# Equivalent Dielectric

Here we use the formula proposed by A.Messiaen et al. (Plasma Phys. Control. Fusion 53 2011 085020), to get a dielectric permittivity $K_D$ somewhat equivalent to a cold plasma of density $n_e$ :
$$
K_D = 1 + \frac{\omega_{pi}^2}{\omega \omega_{ci} + \omega_{ci}^2}
$$

```python
def K_D(n_e=1e18/u.m**3, B_T=3*u.T, f=50e6/u.s):
    omega_pi = plasma_frequency(n_e, particle='D+')
    omega_ci = gyrofrequency(B_T, particle='D+', signed=True)
    omega = 2*np.pi * f * u.rad
    return 1 + omega_pi**2 / (omega*omega_ci + omega_ci**2)
```

```python
n_e = 5e18 /u.m**3 # m^-3
B_T = 3 *u.T # T
f = 50e6 /u.s # Hz

K_D(n_e=n_e, B_T=B_T, f=f)
```

As stated in A. Messiaen et al., ‘Simulation of ICRF antenna plasma loading by a dielectric dummy load. Application to the ITER case’, Fusion Eng. Des., vol. 86, no. 6–8, pp. 855–859, Oct. 2011, the previous formula should be equivalent to 
$$K_D  = S - D$$

```python
S,D,P = cold_plasma_permittivity_SDP(B=B_T, species=['H+', 'D+', 'e'], n=[0.05*n_e, 0.95*n_e, n_e], omega=2*pi*f*u.rad)
print(S)
```

## inhomogeneous profiles

```python
# plasma starts at x=0 with ne0, density increases up to ne1 at x=x_core then constant
ne_0 = 1e15/u.m**3
ne_1 = 45e18/u.m**3
x_plasma = 0 * u.m  # (in HFSS plasma starts at x=0)
x_core = 0.3 * u.m
```

```python
# into Stix parameters
S_0, D_0, P_0 = cold_plasma_permittivity_SDP(B=B_T, species=['H+', 'D+', 'e'], 
                                             n=[0.05*ne_0, 0.95*ne_0, ne_0], omega=2*pi*f*u.rad)
print(S_0, D_0, P_0)

S_1, D_1, P_1 = cold_plasma_permittivity_SDP(B=B_T, species=['H+', 'D+', 'e'], 
                                             n=[0.05*ne_1, 0.95*ne_1, ne_1], omega=2*pi*f*u.rad)
print(S_1, D_1, P_1)

print('V0, gradient, V1')
print('{}, {}, {}'.format(S_0, (S_1 - S_0) / (x_core - x_plasma), S_1 ))
print('{}, {}, {}'.format(D_0, (D_1 - D_0) / (x_core - x_plasma), D_1 ))
print('{}, {}, {}'.format(P_0, (P_1 - P_0) / (x_core - x_plasma), P_1 ))
```

```python
# into equivalent dielectric
K_D0 =  K_D(ne_0)
K_D1 = K_D(ne_1)

gradient = (K_D1 - K_D0) / (x_core - x_plasma)
print(f"K_D1 = {K_D1}")
print(f"gradient={gradient}")
```

## Generate inhomogeneous dielectric profiles for HFSS

```python
# There is three profiles 
# for three Averaged Line Density (LAD)
R1_H, ne1_H = np.loadtxt('IC/Ne_prof_WEST_Hmode_01_LAD6_Rsep_293.txt', skiprows=1, unpack=True)
R2_H, ne2_H = np.loadtxt('IC/Ne_prof_WEST_Hmode_01_LAD9_Rsep_293.txt', skiprows=1, unpack=True)
R3_H, ne3_H = np.loadtxt('IC/Ne_prof_WEST_Hmode_01_LAD12_Rsep_293.txt', skiprows=1, unpack=True)
```

```python
# These profile are defined vs the radial position (in metre)
# In HFSS, the plasma profiles should start a x=0, so we reverse the abscisse
# from a given radius
def x(R, R_x0=2.93):
    return - R1_H + R_x0
```

```python
fig, ax = plt.subplots()
ax.plot(R1_H, ne1_H)
ax.plot(R2_H, ne2_H) # x 1.5
ax.plot(R3_H, ne3_H) # x 2
ax.set_xlabel('R [m]', fontsize=14)
ax.set_ylabel('$n_e$ [$ m^{-3}$ ]')
ax.set_title('ne(x) for R_ant=2.93')
ax.axvline(2.93, color='k', ls='--')

fig.savefig('ne_vs_R.png')
```

```python
plt.subplots()
plt.plot(x(R1_H), ne1_H)
plt.plot(x(R2_H), ne2_H)
plt.plot(x(R3_H), ne3_H)
plt.xlabel('x [m]', fontsize=14)
plt.ylabel('$n_e$ [$ m^{-3}$ ]')
plt.title('ne(x) for R_ant=2.93')
fig.savefig('ne_vs_x.png')
```

```python
# convert these profiles into equivalent dielectric profiles
K_D_LAD6 = K_D(n_e=ne1_H/u.m**3, B_T=3*u.T, f=50e6/u.s)
K_D_LAD9 = K_D(n_e=ne2_H/u.m**3, B_T=3*u.T, f=50e6/u.s)
K_D_LAD12 = K_D(n_e=ne3_H/u.m**3, B_T=3*u.T, f=50e6/u.s)
```

```python
fig, ax = plt.subplots()
ax.plot(x(R1_H), K_D_LAD6)
ax.plot(x(R2_H), K_D_LAD9)
ax.plot(x(R3_H), K_D_LAD12)
```

```python
# reverse the array in order to have increasing values of x for HFSS compatibility
np.savetxt('IC/HFSS_dielectric_profile_LAD6.txt', np.flipud(np.array([x(R1_H), K_D_LAD6]).T))
np.savetxt('IC/HFSS_dielectric_profile_LAD9.txt', np.flipud(np.array([x(R2_H), K_D_LAD9]).T))
np.savetxt('IC/HFSS_dielectric_profile_LAD12.txt', np.flipud(np.array([x(R3_H), K_D_LAD12]).T))
```
